# Conditionals in Python 3

## Boolean data type

A boolean expression is an expression that results in either <i>true</i> or <i>false</i>. In Python 3 Boolean is a standard built-in type (https://docs.python.org/3.4/library/stdtypes.html). Boolean variables take on values of either <i>True</i> or <i>False</i>.

Variables or expressions could be tested to check whether they are true or false. According to python documentation [1] the following cases are considered False:
- None
- False
- zero of any numeric type, for example, 0, 0.0, 0j.
- any empty sequence, for example, '', (), [].
- any empty mapping, for example, {}.
- instances of user-defined classes, if the class defines a __bool__() or __len__() method, when that method returns the integer zero or bool value False.(\*)
In any other case the value of the object would be considered True.
(\*)This would be discussed later in the context of classes.

### Examples

```python
print(bool(1))
print(bool(0))
print(bool(None))
print(bool('should be true'))
```

## Boolean operations

Python 3 defines several boolean operations, which are <i>and, or, not</i>. These are basic boolean operations well described in http://en.wikipedia.org/wiki/Boolean_algebra#Basic_operations. 

### Examples

```python
print(bool(1 and 2))
print(bool('cat' and 'dog'))
print(bool('cat' and 0))
print(bool([] or True))
print(bool(False or True))
print(bool(None or ''))
print(bool(not None))
print(bool(not True))
```

## IF-statement and conditional execution

As in many other programming languages you can execute parts of the code based on a boolean condition. The general syntax for conditional, the if statement, is presented below:

```python
condition1 = True #proxy variable to avoid execution errors
condition2 = True #proxy variable to avoid execution errors
condition3 = True #proxy variable to avoid execution errors

#start conditional execution with first condition
if condition1:
    # if condition1 is True, execute the following code
    # put your code here
    print('just a proxy print');
    
# if condition1 is False, try another condition
# you can have as many elif-statements as needed, or even not have any
elif condition2:
    # if condition1 is False and condition2 is True, execute the following code
    # put your code here
    print('just a proxy print');
    
# if condition1 is False, condition2 is False try another condition
elif condition3:
    # if condition1 is False and condition2 is False and condition3 is True, execute the following code
    # put your code here
    print('just a proxy print');

# if the original condition1 and all the following conditions are False
# you can have only one else-statement, or no else-statements at all
else:
    # if all previous conditions are False, execute the following code
    # put your code here
    print('just a proxy print');
      
    
```

Note that if-statements can be nested, which means that you can put a separate if-statement (along with else-statement and elif-statements, if needed) inside each condition, after "# put your code here" in thetemplate above.

# Recursion

Recursion is widespread programming concept used by many algorithms. See http://en.wikipedia.org/wiki/Recursion_(computer_science) for a general overview of what recursion is. Using recursion is sometimes the only way to solve a computational problem - a simple example are Fibonacci numbers (http://en.wikipedia.org/wiki/Fibonacci_number).

Using recursion in Python is simple, since functions in Python can call themselves:

```python
# define a recursive function
def recursiveFunc(n):
    # let the function call itself based on a condition
    if n > 0:
        recursiveFunc(n-1)
```

Note the condition and argument decrement in the example above. If the function calls itself unconditionally, then there is a chance that the process of recursion would be perpetual. Of course, since each function call takes-up some RAM (memory), such perpetual execution would end - when the memory runs out. To avoid such memory <i>leaks</i> Python has a means of detecting such deep recursion - trying to make too many recursive function calls will result in an error:

```python
def infRecursion():
    infRecursion()
    
infRecursion()
```

For more information on recursion stack and such behavior of perpetual recursion see http://www.greenteapress.com/thinkpython/html/thinkpython006.html#toc59

## Mutual recursion

A special case of recursion called <i>mutual recursion</i> when tow or more functions recursively call each other:

```python
def recursiveFunc1(n):
    if n > 0:
        recursiveFunc2(n-1) #note that we are calling recursiveFunc2 before defining it in the code!
    
def recursiveFunc2(n):
    if n > 0:
        recursiveFunc1(n-1)
    
recursiveFunc1(3) # start recursive execution
```

As we know from previous materials, we should define a function before making calls to it. The example above, although fully working, seems to be breaking this rule. In fact, the code above is correct, because when we start recursive execution (last line), both recursive functions are already defined.
